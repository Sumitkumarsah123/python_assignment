# Qn. 1
A ball is thrown vertically up in the air from a height ho above the ground at an initial velocity vo. Its subsequent height h and velocity ùë£ are given by the equations.
$ h=ho+ vot -1/2gt^2 ,$
$ ùë£ = ùë£o ‚àíùëîùë° $
where ùëî = 9.8 is the acceleration due to gravity in $ùëö/ùë†^2$.
Write a script that finds the height h and velocity ùë£ at a time ùë° after the ball is thrown. Start the code by setting ho = 1.2 (meters) and ùë£o = 5.4 (m/s) and have your code print out the values of height and velocity. Then use the script to find the height and velocity after 0.5 seconds. Then modify your script to find them after 2.0 seconds. 
 
    

# Answer

```python

def height(t, g=9.8,vo=5.4, ho=1.2):# defining a function
    h= ho+ vo*t-1/2*g*t**2
    print("the height at time =" ,h)

def velocity(t, g=9.8, vo=5.4):
    v = vo - g*t
    print("Velocity at time=", v)

# at time 0.5 second   
height(0.5)# calling a function
velocity(0.5)
    
# at time 2.0 second
height(2.0)
velocity(2.0)
```

where ùëî = 9.8 ùëö/ùë†
Suppose ho = 10 ùëö. Find the sequence of times when the ball passes each half meter assuming the ball is dropped at ùë° = 0. Hint: Create a NumPy array for y that goes from 10 to 0 in increments of -0.5 using the arrange function.
Solving the above equation for t, show that
ùë° = $ \frac{\sqrt(2(ho-y))}{g} $ 

(i) Using this equation and the array you created, find the sequence of times when the ball passes each half meter.
(ii) Recalling that the average velocity over an interval Œîùë° is defined as ùë£= $\frac{Œîùë¶}{Œîùë°} $, find the average velocity for each time interval in the problem using NumPy arrays.

```python
Y = np.arange(10, -0.5, -0.5)
Y
```

```python
'''Sequence of times when ball passes each half meter
   Defining time (t) as a function of height (y)'''

def t(y, g=9.8, h0=10):
 return np.sqrt(2 * (h0 - y) / g)

T = t(Y)
T

```

```python
#Finding average velocity (m/s) on above-mentioned intervals
def avg_velocity(T, Y):
    avg = []
    for t, y in zip(T[:-1], Y[:-1]):
      avg.append((y - Y[np.where(Y==y)[0][0] + 1]) / (T[np.where(T==t)[0][0]+1] - t))
    return avg
avg_velocity(T, Y)
```

# Qn. 5

Write a program to tell the nature of the roots and values of the roots of a quadratic equation $ ùëéùë•^2 +ùëèùë• +ùëê = 0,ùëé ‚â† 0 $. 

# Answer

```python
import math
def findroot(a,b,c): #defining the function with parameter a,b and c
    if a==0:
        print("invalid, please enter the value greater than 0")
    d= b**2- 4*a*c
    sqrt_val= math.sqrt(abs(d))
    if d>0:
        print("Roots are real and different ") 
        print((-b + sqrt_val)/(2 * a)) 
        print((-b - sqrt_val)/(2 * a)) 
    elif d == 0: 
        print("Roots are real and same") 
        print(-b / (2*a)) 
    else: 
        print("Roots are complex") 
        print(- b / (2*a) , " + i", sqrt_val) 
        print(- b / (2*a) , " - i", sqrt_val) 
  
findroot(2,-1,2)  # calling the function 
```

# Q.n 13

Write the program for (i) surface plot (ii) contour plot on ùë•ùë¶ ‚àíplane  of the function 
ùëì(ùë•,ùë¶) = $ \frac{1-x/2+ x^5+y^5}{\exp(-x^2-y^2)} $ on the domain $-3<=x<=3 $ and $-3<=y<=3 $.

# Answer

# Contour plot

```python

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
def f(x, y):
     return (1 - (x/2) + x**5 + y**5) / (np.exp((-x**2) - (y**2)))
Z = f(X, Y)

fig = plt.figure(figsize=(10,6))
plt.contourf(X, Y, Z, 10, cmap='magma')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.grid([])
```

# Surface plot

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
def f(x, y):
     return (1 - (x/2) + x**5 + y**5) / (np.exp((-x**2) - (y**2)))
Z = f(X, Y)

fig = plt.figure(figsize=(10,6))
ax = mplot3d.Axes3D(fig);
ax.plot_surface(X, Y, Z, cmap='plasma')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
